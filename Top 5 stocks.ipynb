```python
import yfinance as yf
```

```python
# Process first dataset
tsla_df = yf.download("TSLA",start="2020-12-01",end="2024-12-01")
```

```python
tsla_df.head(2)
```

```python
tsla_df.columns.values
```

```python
tsla_df.columns = [value[0] for value in tsla_df.columns.values]
tsla_df.head(2)
```

```python
tsla_df.insert(0,'Date',tsla_df.index)
tsla_df.reset_index(drop=True,inplace=True)
```

```python
tsla_df.insert(1,'Ticker','TSLA')
```

```python
tsla_df.head(2)
```

```python
# Processing 2nd dataset
smci_df = yf.download("SMCI",start="2020-12-01",end="2024-12-01")
```

```python
smci_df.head(2)
```

```python
smci_df.columns = [value[0] for value in smci_df.columns.values]
smci_df.head(2)
```

```python
smci_df.insert(0, 'Date',smci_df.index)
smci_df.insert(1,'Ticker','SMCI')
smci_df.reset_index(drop=True,inplace=True)
```

```python
import pandas as pd
```

```python
merged_df = pd.concat([tsla_df,smci_df,nvda_df,intc_df,pltr_df])
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df["Date"] = pd.to_datetime(merged_df["Date"], format="%Y-%m-%d %H:%M:%S%z").dt.date
merged_df["Date"] = pd.to_datetime(merged_df["Date"])
```

```python
merged_df.info()
```

```python
merged_df.head()
```

```python
merged_df['Ticker'].unique()
```

```python
import plotly.express as px
```

Each shares "Open values" as per date

```python
# This is 4 years stock of Tesla Inc. From this diagram, the open price fluctuates over time, with a notable peak in November 2021.
x = px.line(merged_df[merged_df['Ticker']=='TSLA'], x='Date', y='Open')
x.update_layout(title='Tesla Inc', title_x=0.5)
x.show()
```

```python
# This is a 4-year stock chart of Supermicro. On 2021, it is very low. but, Starting in January 2024, the stock begins to rise.
px.area(merged_df[merged_df['Ticker']=='SMCI'], x='Date', y='Open', title='Super Micro Computer, Inc.')
```

```python
# This is a 4-year stock chart of Nvidia, The open price is rising slowly
px.line(merged_df[merged_df['Ticker']=='NVDA'], x='Date', y='Open', title='NVIDIA Corporation')
```

```python
# This is a 4-year stock chart of Intel. Initially, the opening stock price is high, but after that, the stock begins to fall.
px.line(merged_df[merged_df['Ticker']=='INTC'], x='Date', y='Open', title='Intel Corporation')
```

```python
# This is a 4-year stock chart of Palantir Technologies. The opening stock price fluctuated.
px.area(merged_df[merged_df['Ticker']=='PLTR'], x='Date', y='Open', title='Palantir Technologies Inc')
```

```python
# In this chart, shows the open price of each stock as per date from "2020-12-01" to "2024-12-01")
px.bar(merged_df, y='Ticker', x='Open', title='Top 4 stocks Opening Information of past 4 years', hover_name='Ticker',range_x=[0,merged_df['Open'].max()], color='Open', animation_frame='Date')
```
