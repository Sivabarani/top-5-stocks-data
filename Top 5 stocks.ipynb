```python
import plotly.express as px
```

Each shares "Open values" as per date

```python
# This is 5-year stock chart of Tesla Inc. 
# From this diagram, the open price fluctuates over time, with a notable peak in November 2021.
# X axis represents Date, Y Axis Represents Open
tsla_chart = px.line(merged_df[merged_df['Ticker']=='TSLA'], x='Date', y='Open')
tsla_chart.update_layout(title='Tesla Inc', title_x=0.5)
tsla_chart.show()
```

```python
# This is a 5-year stock chart of Supermicro. 
# On 2019, it is very low. but, Starting in January 2024, the stock begins to rise.
# X axis represents Date, Y Axis Represents Open
smci_chart = px.area(merged_df[merged_df['Ticker']=='SMCI'], x='Date', y='Open', title='Super Micro Computer, Inc.')
smci_chart.update_layout(title='Super Micro Computer, Inc.', title_x=0.5)
smci_chart.show()
```

```python
# This is a 5-year stock chart of Nvidia, 
# The open price is rising slowly and reach the peak on Jun 2024
# X axis represents Date, Y Axis Represents Open
nvda_chart = px.line(merged_df[merged_df['Ticker']=='NVDA'], x='Date', y='Open', title='NVIDIA Corporation')
nvda_chart.update_layout(title='NVIDIA Corporation', title_x=0.5)
nvda_chart.show()
```

```python
# This is a 5-year stock chart of Intel. 
# Initially, the opening stock price is high, but after that, the stock begins to little fall compared to previous year.
# X axis represents Date, Y Axis Represents Open
intc_chart = px.area(merged_df[merged_df['Ticker']=='INTC'], x='Date', y='Open', title='Intel Corporation')
intc_chart.update_layout(title='Intel Corporation', title_x=0.5)
intc_chart.show()
```

```python
# This is a 5-year stock chart of Palantir Technologies. 
# The opening stock price fluctuated. On Start of 2021, it little rised, then fall, on june 2024 it begins to rise
# X axis represents Date, Y Axis Represents Open
pltr_chart = px.area(merged_df[merged_df['Ticker']=='PLTR'], x='Date', y='Open', title='Palantir Technologies Inc')
pltr_chart.update_layout(title='Palantir Technologies Inc', title_x=0.5)
pltr_chart.show()
```

```python
# This chart shows the opening price of each stock for each date from "2019-12-01" to "2024-12-01"
# It visualizes the opening price of stocks over time, highlighting the variance in stock prices.
# The data is animated, with a bar chart showing stock opening prices by date.
px.bar(merged_df, y='Ticker', x='Open', title='Top 4 stocks Opening Information of past 5 years', hover_name='Ticker',range_x=[0,merged_df['Open'].max()], color='Open', animation_frame='Date')
```

```python
# This chart shows the closing price of each stock for each date from "2019-12-01" to "2024-12-01"
# It visualizes the closing price of stocks over time, highlighting the variance in stock prices.
# The data is animated, with a bar chart showing stock closing prices by date.
px.bar(merged_df, y='Ticker', x='Close', title='Top 5 stocks Closing Information of past 4 years', hover_name='Ticker',range_x=[0,merged_df['Close'].max()], color='Close', animation_frame='Date')
```

```python
top_open = merged_df.groupby('Ticker')['Open'].max().reset_index()
top_open
```

```python
# Create a pie chart to visualize the maximum opening prices of the top stocks over the past 5 years
# 'values' represent the maximum opening price ('Open') for each stock
# 'names' represent the stock ticker names ('Ticker')

"""From this chart, we can observe that Tesla Inc. (TSLA) occupies half of the pie, 
indicating that Tesla has the highest opening price over the past 5 years 
compared to the other stocks in the chart. This suggests that Tesla's stock 
price has seen significant growth during this period, which is visually represented 
by the large share of the pie chart dedicated to TSLA.
On the other hand, Intel (INTC) occupies the smallest slice, indicating that 
its opening price has been the lowest among the selected stocks during the same period."""
stockes_pie_chart = px.pie(top_open, values='Open', names='Ticker', title='Top 5 stocks open of past 5 years' )
stockes_pie_chart.update_traces(textposition="inside",textinfo="label+text+value+percent")
stockes_pie_chart.show()
```

```python
copy_merged_df = merged_df.copy()
copy_merged_df['Year'] = copy_merged_df['Date'].dt.year
copy_merged_df.head()
```

```python
top_open_by_year = copy_merged_df.groupby(['Year','Ticker'])['Open'].max().reset_index()
top_open_by_year.head()
```

```python
# This chart represents the maximum opening stock prices over the past 5 years for each stock ticker.
"""Tesla (TSLA) consistently ranks in the top position each year, with the highest opening prices in most years.
By interacting with the chart and selecting specific years, we can see the maximum opening price for each stock during that year.
In 2019, only Intel (INTC) took the first position, with Tesla in second place. 
However, from 2020 to 2023, Tesla held the top position.
Intel (INTC) frequently ranks in second place in most years, but in 2024, it falls to the last position, 
with NVIDIA (NVDA) taking second place instead.
Stocks like SMCC and PLTR show much lower maximum opening prices compared to the others, reflecting their relatively lower market performance."""
top_open_by_year_chart = px.sunburst(top_open_by_year, path=['Year','Ticker'], values='Open')
top_open_by_year_chart.update_layout(title='Maximum stocks open of past 5 years for each Ticker', title_x=0.5)
top_open_by_year_chart.show()
```

```python
low_open_by_year = copy_merged_df.groupby(['Year','Ticker'])['Open'].min().reset_index()
low_open_by_year.head()
```

```python
# This chart represents the minimum opening stock prices over the past 5 years for each stock ticker.
# SMCI (Super Micro Computer) had the lowest opening price in each year, taking the last position in the chart.
# On the other hand, Tesla (TSLA) consistently ranks first, having one of the highest minimum opening prices in most years.
# The sunburst chart visualizes the minimum opening price for each stock across the years.
# By interacting with the chart, we can see how each stock performed in terms of the lowest opening prices by year.s
low_open_by_year_chart = px.sunburst(low_open_by_year, path=['Year','Ticker'], values='Open')
low_open_by_year_chart.update_layout(title='Minimum stocks open of past 5 years for each Ticker', title_x=0.5)
low_open_by_year_chart.show()
```
